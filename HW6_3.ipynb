```python
import pandas as pd
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
world_cup = pd.read_csv('/content/drive/MyDrive/DATA/WorldCupMatches.csv')
world_cup.head()
```

```python
world_cupdf = world_cup.groupby(['Home Team Initials','Away Team Initials']).size().reset_index(name='weight')
world_cupdf.head(15)
```

```python
HomeGoalsSum = world_cup.groupby(['Home Team Initials','Away Team Initials']).sum(['Home Team Goals']).reset_index()
HomeGoalsSum.head(15)
```

```python
team_pairs = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).size().reset_index(name='weight')
team_pairs['Home Goal Sums'] = HomeGoalSum['Home Team Goals']

team_pairs.head(15)
```

```python
team_pairs.to_csv('/content/drive/MyDrive/Colab Notebooks/team_pairs.csv')
```
